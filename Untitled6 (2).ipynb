1. PassengerId  : 승객 고유 ID
2. Survived     : 생존 여부 (0 = 사망, 1 = 생존)
3. Pclass       : 좌석 등급 (1 = 1등석, 2 = 2등석, 3 = 3등석)
4. Name         : 이름
5. Sex          : 성별
6. Age          : 나이
7. SibSp        : 형제/배우자 수
8. Parch        : 부모/자녀 수
9. Ticket       : 티켓 번호
10. Fare        : 운임
11. Cabin       : 객실 번호
12. Embarked    : 탑승 항구 (C = Cherbourg, Q = Queenstown, S = Southampton)

###결측치 확인 및 처리

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 결측치 처리 예시
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df.drop(columns=['Cabin'], inplace=True)  # Cabin은 결측치가 많아 제거
```

### 데이터분석

```python
# 생존자 수 시각화 (사망자 vs 생존자 수)
sns.countplot(data=df, x='Survived')
plt.title("생존자 수 (0=사망, 1=생존)")
plt.show()
```

```python
# 생존자 비율
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
# 객실 등급(Pclass)에 따른 생존율 분석 (1등석 ~ 3등석)
sns.barplot(x='Sex', y='Survived', data=df)  # 등급별 생존율 비교
plt.title('Survival Rate by Sex')
plt.show()
```

```python
#객실 등급(Pclass)에 따른 생존율 분석 (1등석 ~ 3등석)
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Class') # 등급별 생존율 비교
plt.show()
```

```python
#나이에 따른 생존율 분석
sns.histplot(data=df, x='Age', hue='Survived', bins=20, kde=True)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
#가족 동반자 수 (SibSp + Parch)와 생존율
df['FamilySize'] = df['SibSp'] + df['Parch']
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
#탑승지에 따른 생존율
sns.countplot(data=df, x='Embarked', hue='Survived')
plt.title("탑승지에 따른 생존 비교")
plt.show()
```
